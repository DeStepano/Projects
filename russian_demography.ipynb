# Анализ демографии России
Цель: 
1. Проанализировать данных демографии России с 1990 по 2017 годы.
2. Сделать выводы.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
dem = pd.read_csv('russian_demography.csv')
dem_g = dem.pivot_table(["npg", "birth_rate","death_rate","gdw", "urbanization"], "year", aggfunc=np.mean)
dem_g
```

```python
dem
```

```python
dem_g.describe()
```

```python
ax = dem_g.plot(y="npg", xlim=(1990,2017), ylim=(-6, 6), figsize=(8,5), linewidth=3)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.tick_params(bottom=False, left=False)
```

## Вывод: 
Прирост населения значительно падал с 1990 года в России (с 4 человек на 1000 до -5 человек на 1000), и к 2017 году он был всё ещё отрицательным. 
Прирост населения (npg) - Количество новорожденных за вычетом количество умерших на 1000 человек.

```python
ax = dem_g.plot(y="urbanization", xlim=(1990,2017), ylim=(68.5, 70.5), figsize=(8,5), linewidth=3)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.tick_params(bottom=False, left=False)
```

## Вывод: 
Уровень урбанизации населения переменно растет и достигает 70.25% к 2017 году.

```python
ax = dem_g.plot(y="gdw", xlim=(1990,2017), ylim=(58, 78), figsize=(8,5), linewidth=3)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.tick_params(bottom=False, left=False)
```

"gdw" - количество нетрудоспособных людей на 1000 человек.
## Вывод: 
Видно, что gwd падал до 2007 года, а потом начал расти. Это может свидетельствовать о изменении среднего возраста населения [среднего возраста населения](https://thebell.io/mir-stremitelno-stareet-i-rossiya-vmeste-s-nim-chto-eto-znachit).

```python
ax = dem_g.plot(y="death_rate", xlim=(1990,2017), ylim=(9, 78), figsize=(8,5), linewidth=3)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.tick_params(bottom=False, left=False)
```

Количество смертей на 1000 человек почти не изменялось за 27 лет.

```python
fig = plt.figure(figsize=(16,13))
ax = fig.add_subplot(3,1,1)
ax2 = fig.add_subplot(3,1,2)
#ax3 = fig.add_subplot(3,1,3)
sc = ax.scatter(dem['year'],  dem['npg'], label=None,
            c=dem['urbanization'], cmap='viridis',
            linewidth=0, alpha=0.7)
ax.set_xticks([])
ax.set_ylabel('npg')
cbar = fig.colorbar(sc, ax=(ax,ax2))
cbar.set_label('% of urban population')
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.tick_params(bottom=False, left=False)

ax2.scatter(dem["year"], dem["death_rate"], c=dem["urbanization"], cmap="viridis", linewidth=0, alpha=0.7)
ax2.set_xlabel("year")
ax2.set_ylabel("death rate")
ax2.spines["right"].set_visible(False)
ax2.spines["left"].set_visible(False)
ax2.spines["top"].set_visible(False)
ax2.spines["bottom"].set_visible(False)
ax2.tick_params(bottom=False, left=False)
fig2 = plt.figure(figsize=(8,10))
ax2_1 = fig2.add_subplot(2,1,1)
ax2_2 = fig2.add_subplot(2,1,2)
ax2_1.scatter(dem["urbanization"],dem["death_rate"], linewidth=0)
ax2_2.scatter(dem["urbanization"], dem["birth_rate"], linewidth=0)
ax2_1.set_xticks([])
ax2_1.set_xlim(0,100)
for k,c in ax2_1.spines.items():
    c.set_visible(False)
for k,c in ax2_2.spines.items():
    c.set_visible(False)
ax2_1.tick_params(left=False)
ax2_2.set_xlim(0,100)
ax2_2.tick_params(bottom=False, left=False)
ax2_2.set_xlabel("% of urban population")
ax2_1.set_ylabel("death rate")
ax2_2.set_ylabel("birth rate")
```

## Вывод
Количество новорожденных за год коррелирует с процентом урбанизации так, что чем больше уровень урбанизации, тем реже люди продлевают свой род, от этого рост населения замедляется, в некоторых случаях становится отрицательным. Научно это можно оправдать биологией - для природы, эволюции и для нашего организма мы всего лишь очередные носители гена, и наша главная цель - продолжение рода, поэтому когда возникает какая-либо угроза жизни человека, например война, желание оставить свои гены (размножиться) возрастает. Чем выше уровень жизни, тем безопаснее себя чувствует человек. Видно, что в развитых странах, где преобладают ценности самореализации, а не выживания, население практически не растет, в то время как в слаборазвитых странах население активно растет.

```python
npg_mean = dem.pivot_table("npg", "year")
sam_mosc = dem[(dem["region"]=="Samara Oblast")|(dem["region"]=="Moscow")]
ax = sam_mosc[sam_mosc["region"]=="Moscow"].plot("year", "npg", label="Moscow", figsize=(8,5), linewidth=3)
ax.plot(sam_mosc[sam_mosc["region"]== "Samara Oblast"]["year"], sam_mosc[sam_mosc["region"]== "Samara Oblast"]["npg"], linewidth=3)
ax.plot(npg_mean, linewidth=3)
ax.legend(["Moscow", "Samara Oblast", "npg mean"])
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.spines["bottom"].set_visible(False)
ax.set_ylabel("npg")
ax.set_xlim(1990,2017)
ax.set_ylim(-10,2)
ax.tick_params(bottom=False, left=False)
```
