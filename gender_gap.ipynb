# Гендерный разрыв в профессиях 
Проанализируем процентное соотношение количества дипломов различных факультетов разных полов

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
```

```python
fig = plt.figure(figsize=(14,25))
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
cb_orange = (255/255, 128/255, 14/255)
cb_grey = (89/255,89/255,89/255)
stem_index = np.arange(1,17,3)
lib_index = np.arange(2,15,3)
oth_index = np.arange(3,19,3)
print(stem_index, lib_index, oth_index)
for sp in range(0,6):
    ax = fig.add_subplot(6,3,stem_index[sp])
    ax.plot(women_degrees["Year"], women_degrees[stem_cats[sp]], label="Women", linewidth=3, c=cb_orange)
    ax.plot(women_degrees["Year"], 100 - women_degrees[stem_cats[sp]], label="Men", linewidth=3, c=cb_grey)
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.set_ylim(0,100)
    ax.set_xlim(1968,2011)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom=False,left=False)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    if sp == 0:
        ax.text(2004, 80, "Women")
        ax.text(2007, 16, "Men")
    if sp != 5:
        ax.tick_params(labelbottom = False)
for sp in range(0,5):
    ax = fig.add_subplot(6, 3,lib_index[sp])
    ax.plot(women_degrees["Year"], women_degrees[lib_arts_cats[sp]], label="Women", linewidth=3, c=cb_orange)
    ax.plot(women_degrees['Year'], 100 - women_degrees[lib_arts_cats[sp]], label='Men', linewidth=3, c=cb_grey)
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.set_ylim(0,100)
    ax.set_xlim(1968,2011)
    ax.set_title(lib_arts_cats[sp])
    ax.tick_params(bottom=False, left=False)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    if sp == 0:
        ax.text(2004,75,"Women")
        ax.text(2007,23,"Men")
    if sp != 4:
        ax.tick_params(labelbottom = False)

for sp in range(0,6):
    ax = fig.add_subplot(6, 3,oth_index[sp])
    ax.plot(women_degrees["Year"], women_degrees[other_cats[sp]], label="Women", linewidth=3, c=cb_orange)
    ax.plot(women_degrees['Year'], 100 - women_degrees[other_cats[sp]], label='Men', linewidth=3, c=cb_grey)
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.set_ylim(0,100)
    ax.set_xlim(1968,2011)
    ax.set_title(other_cats[sp])
    ax.tick_params(bottom=False, left=False)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    if sp == 0:
        ax.text(2004,90,"Women")
        ax.text(2007,8,"Men")
    if sp != 5:
        ax.tick_params(labelbottom = False)
fig.savefig('dada2.png')
plt.show()
```
