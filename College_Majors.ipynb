# Управляемый проект: Визуализация доходов на основе специальностей колледжа
Dataset состоит из столбцов:
* Rank - Рейтинг по среднему доходу (набор данных упорядочен по этому столбцу).
* Major_code - Основной код.
* Major - Спецальность.
* Major_category - Направление спецальности.
* Total - Общее число людей, получивших специалитет.
* Sample_size - Число занятых людей.
* Men - Количество мужчин.
* Women - количество женщин.
* ShareWomen - Отношение количества женщин к общему количеству людей.
* Employed - Количество занятых.
* Median - Средняя зарплата за год.
* Low_wage_jobs - Количество низкооплачиваемых рабочих мест в сфере обслуживания.
* Full_time - Количество работающих 35 часов и более.
* Part_time - Количество работающих менее 35 часов.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas.plotting import scatter_matrix
%matplotlib inline
recent_grads = pd.read_csv('recent-grads.csv')
print(recent_grads.iloc[1])
print(recent_grads.head())
print(recent_grads.tail())
print(recent_grads.describe())
raw_data_count = recent_grads.shape[0]
recent_grads = recent_grads.dropna(axis=0)
cleaned_data_count = recent_grads.shape[0] 
print(raw_data_count, cleaned_data_count)
recent_grads
```

Видно, что есть такие направления, как образование, здравоохранение, социальные работы, психология. где женщин намного больше, чем мужчин, также есть и специальности, в которых мужчин значительно больше: инженерия, математика и компьютеры, и направления, в которых количество мужчин и женщин примерно одинаковое: право и политика, физические науки.

```python
fig = plt.figure(figsize=(8,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax1.boxplot(recent_grads["Median"], showmeans=True)
ax1.set_xticklabels(["Median"])
ticksy = range(20000, 110001, 10000)
ax1.set_yticks(ticksy)
ax2.boxplot(recent_grads["Unemployment_rate"], showmeans=True)
ax2.set_xticklabels(["Unemployment_rate"])
print(recent_grads["Median"].median())
recent_grads[["Median", "Unemployment_rate"]].describe()
```

Медиана на 36000 среднего заработка. 25% заработных плат находятся в диапозоне от 34000 до 22000. 50% з/п от 34000 до 45000. 25% от 45000 до 60000. Также есть 5 сильных выбросов за квартили, в других случаях мы бы рассмотрели эти специальности подробнее, но здесь мы понимаем, что это вызвано дефицитом работников в этой сфере.
Распределение уровня безработицы кажется намного равномернее, хотя мы также имеем выбросы, что опять же могло потребовать отдельного внимания в других ситуациях, но в данном случае мы понимаем, что это может быть вызвано ситуацией на рынке труда.
