# Анализ комментариев на сайте Hacker News

На сайте Hacker News представлены различные публикации, связанный с IT сферой и не только. 
Мы проанализируем, к каким постам и в какое время пользователи оставляли комментарии.
Для начала просто откроем архив данных и конвертируем его в список списков.

```python
from csv import reader
hn = list(reader(open('hacker_news.csv')))
```

# Введение
Далее мы просто проанализируем, какие строки и столбцы мы имеем.

```python
print(hn[:5])
```

# Заголовки и основная часть данных
Для нашего же удобста в анализе данных мы разделим заголовоки и нужные нам данные. 

```python
headers = hn[0]
hn = hn[1:]
print(headers, hn[:5])
```

# Разделение постов на три класса
Классы:

    1. "Ask Posts" - посты с вопросами от пользователей. 
    Пример: "Can anyone suggest a good RSS newsreader with a set of tech news feeds?".
    
    2. "Show posts" - публикации с проектами пользователей, публикации о новый продуках или просто что-то интересное.
    Пример: "Space Shooter Retro Game Recreated Using Python".
    
    3. "Other Posts" - этот класс включает в себя всё остальное, этот класс нас не особо интересует так, как в нашем проекте мы будем рассматривать два других класса/категории постов
    Пример: "Hedge Fund Star: We Are Under Assault".

```python
ask_posts, show_posts, other_posts = [],[],[]
for row in hn:
    title = row[1].lower()
    if title.startswith('ask hn'):
        ask_posts.append(row)
    elif title.startswith('show hn'):
        show_posts.append(row)
    else:
        other_posts.append(row)

print('The number of ask posts:', len(ask_posts))
print('The number of show posts:', len(show_posts))
print('The number of other posts:', len(other_posts))
```

# Количество комментариев
Теперь нам нужно подсчитать среднее количество комментариев под постами каждого класса/категории. Чтобы было проще находить среднее значение, мы создадим ниже функцию, которая получает на вход список спиков и нужный столбец для подсчёта среднего значения. 

```python
def avg_count(list0, column):
    total_count = 0 
    for row in list0: 
        num = int(row[column])
        total_count += num
    return total_count / len(list0)
        
```

Ниже мы применим нашу новую функцию к нашим данным, разделённым на два класса.

```python
avg_ask_comments = avg_count(ask_posts, 4)
avg_show_comments = avg_count(show_posts, 4)
print("Average number of ask comments:", avg_ask_comments)
print("Average number of show comments:", avg_show_comments)
```

# Вывод
Мы можем сделать предварительный вывод, что публикации с вопросами вызывают больше ответных действий со стороны пользователей. Это можно обосновать тем, что посты с вопросами нуждаются в отклике аудитории, в ответе, а публикации, рассказывающие о новинках, проектах и так далее в этом не нуждаются. 

Поскольку публикации с вопросами вызывают больший ажиотаж, мы сосредоточимся именно на них. 

# Нахождение количества публикаций и общего количества комментариев в каждом часе суток
Ниже мы создадим два словаря, которые будут состоять из информации о количестве публикаций в час и количестве комментариев в час.

```python
import datetime as dt
result_list = []
counts_by_hour = {}
comments_by_hour = {}

for row in ask_posts:
    result_list.append([row[6], int(row[4])])

for row in result_list:
    date = row[0]
    count_com = row[1]
    hour = dt.datetime.strptime(date, "%m/%d/%Y %H:%M").strftime("%H")
    if hour in counts_by_hour:
        counts_by_hour[hour] += 1 
        comments_by_hour[hour] += count_com
    else:
        counts_by_hour[hour] = 1
        comments_by_hour[hour] = count_com
```

# Поиск среднего количества комментариев в каждом часе суток
Теперь нам нужно поделить количество комментариев в часе на количество постов в этом же часе.

```python
avg_by_hour = []
for hour, count in comments_by_hour.items():
    avg_by_hour.append([hour, count/counts_by_hour[hour]])
avg_by_hour = sorted(avg_by_hour)
print(avg_by_hour)
```

# Сортировка данных от наибольшего к наименьшему
Ниже мы просто сортируем наши данные по убыванию и выводим их в удобном для чтения формате.

```python
swap_avg_by_hour = []
for row in avg_by_hour:
    swap_avg_by_hour.append([row[1], row[0]])
sorted_swap = sorted(swap_avg_by_hour, reverse = True)
print("Top 5 Hours for Ask Posts Comments:")
template = "{}: {:.2f} average comments per post"
for c, h in sorted_swap[:5]:
    print(template.format(dt.datetime.strptime(h, "%H").strftime("%H:%M"), c))
```

# Вывод
В конце концов, исходя из полученных нами данных, мы можем сделать вывод, что в среднем посты с вопросами, опубликованные в промежуток с 15:00 до 16:00, получают наибольшее количество комментариев - 38.59. Разница между часами с наибольшим и вторым по величине средним числом комментариев состаляет 60%. Важно заметить, что речь идет о Восточно-Американском часовом поясе (UTC-5). Также нельзя не отметить, что в нашем анализе мы учитывали только публикации, под которыми был оставлен хотя бы один комментарий, то есть посты без комментариев не учитывались. Исходя из этого, стоит поправить наши слова, сказав, что среди постов с комментариями наибольшее среднее количество комментариев имеют публикации с вопросами, опубликованные в промежуток между 15:00 и 16:00.

# Дополнительное задание


Создадим функцию "avg_count_per_hour", которая на вход получает основной список, индекс столбца с нужным нам для анализа значением, индекс столбца с датой публикации поста, название класса/категории данных, названия столбца со значениями, значение True или False, позволяющее вывести отсортированные данные на экран, и также значение True или False относительно вывода среднего значения в каждом посту, а на выход функция даёт нам список списков со среднеми значениями и соответствующими часами.

```python
def avg_count_per_hour(input_list, c_index, t_index, name_of_column="", name_of_class="", print_avg_per_hour=False, print_avg_count=False):
    dict_of_post = {}
    dict_of_counts = {}
    list_of_counts = []
    for row in input_list:
        list_of_counts.append([row[t_index], int(row[c_index])])
    for h, c in list_of_counts:
        h = dt.datetime.strptime(h, "%m/%d/%Y %H:%M").strftime("%H")
        if h in dict_of_post:
            dict_of_post[h] += 1
            dict_of_counts[h] += c
        else:
            dict_of_post[h] = 1 
            dict_of_counts[h] = c
    
    avg_list_of_counts = []
    for h, c in dict_of_counts.items():
        avg_list_of_counts.append([c/dict_of_post[h], h])
    avg_list_of_counts = sorted(avg_list_of_counts, reverse=True)
    if print_avg_count:
        print("The average number of", name_of_column, "per post:")
        avg = 0 
        for h, c in list_of_counts:
            avg += c
        print(avg/len(input_list))
    if print_avg_per_hour:
        print("Top 5 for", name_of_class, name_of_column + ":")
        for c, h in avg_list_of_counts[:5]:
            template = "{}: {:.2f} average {} per hour"
            print(template.format(dt.datetime.strptime(h, "%H").strftime("%H:%M"), c, name_of_column))
    return avg_list_of_counts
```

```python
avg_count_per_hour(ask_posts, 3, 6, "points", "Ask Posts", True, True)
avg_count_per_hour(ask_posts, 4, 6, "comments", "Ask Posts", True, True)
print("")
avg_count_per_hour(show_posts, 3, 6, "points", "Show Posts", True, True)
avg_count_per_hour(show_posts, 4, 6, "comments", "Show Posts", True, True)
print("")
avg_count_per_hour(other_posts, 3, 6, "points", "Other Posts", True, True)
avg_count_per_hour(other_posts, 4, 6, "comments", "Other Posts", True, True)
pass
```
